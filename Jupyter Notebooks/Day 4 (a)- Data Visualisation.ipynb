## Line Plot

### Plotting, Giving Title and Setting Limit

```python
# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset

phool=sns.load_dataset("iris")
phool
# draw a plot say line plot
sns.lineplot(x="petal_length", y= "petal_width", data=phool)
plt.title("Phoolon ka Plot") # setting title
plt.xlim(2) # setting limits
plt.ylim()
plt.show()
```

### Setting Style

```python
# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
phool=sns.load_dataset("iris")
phool
# draw a plot say line plot
sns.set_style("darkgrid")
sns.lineplot(x="petal_length", y= "petal_width", data=phool)
plt.title("Phoolon ka Plot")
plt.show()
```

### Size of Figure

```python
# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
phool=sns.load_dataset("iris")
phool
# draw a plot say line plot 
## setting the theme
plt.figure(figsize=(4,3)) # setting figure size
sns.set_style("darkgrid") # setting tyle
sns.lineplot(x="petal_length", y= "petal_width", data=phool)
plt.title("Phoolon ka Plot")
plt.show()
```

## Bar Plot

```python
# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
phool=sns.load_dataset("iris")
phool
# draw a plot say line plot 
## setting the theme
plt.figure(figsize=(5,5)) # setting figure size
sns.set_style("darkgrid") # setting tyle
sns.barplot(x="species", y= "petal_width", data=phool)
plt.title("Phoolon ka Plot")
plt.show()
```

### Order and Color

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
beri=sns.load_dataset("titanic")
beri
# draw a plot say line plot 
## setting the theme
plt.figure(figsize=(5,5)) # setting figure size
sns.set_style("darkgrid") # setting tyle
sns.barplot(x="sex", y="alone", hue="who", data=beri, order=["female", "male"], color=("salmon")) # giving ordrr and color
plt.title("Titanic Plot")
plt.show()
```

### Removinf Error Bar (Confidential Interval) and Giving Palette

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
beri=sns.load_dataset("titanic")
beri
# draw a plot say line plot 
## setting the theme
plt.figure(figsize=(5,5)) # setting figure size
sns.set_style("darkgrid") # setting tyle
sns.barplot(x="sex", y="alone", hue="who", data=beri, order=["female", "male"], color=("salmon"), ci=None, palette="pastel")
plt.title("Titanic Plot")
plt.show()
```

### Importing Numpy and Estimating operations as mean

```python
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt
# Load dataset
beri=sns.load_dataset("titanic")
beri
# draw a plot say line plot 
## setting the theme
plt.figure(figsize=(5,5)) # setting figure size
sns.set_style("darkgrid") # setting tyle
sns.barplot(x="class", y="fare", hue="sex", data=beri, color=("salmon"), ci=None, palette="pastel", saturation=0, estimator=mean)
plt.title("Titanic Plot")
plt.show()
```

```python
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt
# Load dataset
kashti=sns.load_dataset("titanic")
kashti
# draw a plot say line plot 
## setting the theme
plt.figure(figsize=(5,5)) # setting figure size
sns.set_style("darkgrid") # setting tyle
sns.barplot(x="class", y="fare", hue="sex", data=kashti, 
            linewidth=1, facecolor=(0, 0.25, 0.25, 0.5), 
            errcolor="0.5", edgecolor="0")
plt.title("Titanic Plot")
plt.show()
```
