# Various Plots in Python

```python
import pandas as pd
import seaborn as sns
import numpy
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="darkgrid")
data=pd.read_excel('Cleaned_Dataset.xlsx')
```

```python
data.head(3)
```

```python
# plotting both distibutions on the same figure
fig = sns.kdeplot(data['Light kitni der band hti hy? int'], shade=True, color="r")
fig = sns.kdeplot(data['How many hours you code a day? (int) e.g: 5,4,3'], shade=True, color="b")
plt.title ("Phython_ka_Chilla_Data", size= 16, weight="bold")
plt.ylabel("Density", size= 14, weight="bold")
plt.legend ("LC")
plt.xlabel("Data", size= 14, weight="bold")
plt.show()
```

```python
# plotting both distibutions on the same figure
sns.violinplot(x="Light kitni der band hti hy? int", y="Gender", data=data)
plt.title ("Phython_ka_Chilla_Data", size= 16, weight="bold", y=1.3)
plt.ylabel("Gender", size= 14, weight="bold")
plt.xlabel("Loadshedding (Hours)", size= 14, weight="bold")
plt.show()
```

```python
sns.jointplot(x="Light kitni der band hti hy? int", y="How many hours you code a day? (int) e.g: 5,4,3", data=data, color= "r", kind='scatter')
plt.title ("Phython_ka_Chilla_Data", size= 16, weight="bold", y=1.3)
plt.ylabel("Loadshedding (Hourse)", size= 14, weight="bold")
plt.xlabel("Codding (Hours)", size= 14, weight="bold")
plt.show()
```

```python
sns.jointplot(x=data["Light kitni der band hti hy? int"], y=data["How many hours you code a day? (int) e.g: 5,4,3"], kind='hex')
plt.title ("Phython_ka_Chilla_Data", size= 16, weight="bold", y=1.3)
plt.ylabel("Codding (Hours)", size= 14, weight="bold")
plt.xlabel("Loadshedding (Hours)", size= 14, weight="bold")
plt.show()
```

```python
sns.jointplot(x=data["Light kitni der band hti hy? int"], y=data["How many hours you code a day? (int) e.g: 5,4,3"], color= "salmon", kind='kde')
plt.title ("Phython_ka_Chilla_Data", size= 16, weight="bold", y=1.3)
plt.ylabel("Codding (Hours)", size= 14, weight="bold")
plt.xlabel("Loadshedding (Hours)", size= 14, weight="bold")
plt.show()
```

```python
sns.jointplot(x=data["Light kitni der band hti hy? int"], y=data["How many hours you code a day? (int) e.g: 5,4,3"], color="y", kind='reg')
plt.title ("Phython_ka_Chilla_Data", size= 16, weight="bold", y=1.3)
plt.ylabel("Codding (Hours)", size= 14, weight="bold")
plt.xlabel("Loadshedding (Hours)", size= 14, weight="bold")
plt.show()
```
